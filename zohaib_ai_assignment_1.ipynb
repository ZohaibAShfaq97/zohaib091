```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Most Popular Programming Languages (1).csv')
```

```python
data.describe(include = 'all').T
```

```python
sns.pairplot(data)
plt.show()
```

```python
for i in data.select_dtypes(include = 'number').columns:
    fig = px.histogram(data, x = i)
    fig.update_layout(
    height = 400,
    width = 600,
    title = f'{i}'
    )
    fig.show()
```

```python
line_fig = px.line(data, x = 'Month',
                   y = ['Python Worldwide(%)', 'JavaScript Worldwide(%)', 'Java Worldwide(%)',
                      'C# Worldwide(%)', 'PhP Worldwide(%)', 'Flutter Worldwide(%)',
                      'React Worldwide(%)', 'Swift Worldwide(%)',
                      'TypeScript Worldwide(%)', 'Matlab Worldwide(%)'],
                   title = 'Worldwide Popularity of Programming Languages Over Time')
line_fig.update_layout(
height = 600,
width = 900
)
line_fig.show()
```

```python
bar_fig = px.bar(data.mean(numeric_only = True).reset_index(),
                 x = 'index', y = 0,
                 title = 'Average Worldwide Popularity of Programming Languages',
                 labels = {'index': 'Programming Language', 0: 'Average Percentage (%)'})

bar_fig.show()
```

```python
violin_fig = px.violin(data.melt(id_vars = 'Month',
                                 var_name = 'Language',
                                 value_name = 'Popularity'),

                       x = 'Language',
                       y = 'Popularity',
                       title = 'Distribution of Programming Languages Worldwide Popularity')

violin_fig.show()
```

```python
scatter_fig = px.scatter(data,
                         x = 'Python Worldwide(%)',
                         y = 'JavaScript Worldwide(%)',
                         title = 'Python vs JavaScript Worldwide Popularity')

scatter_fig.show()
```

```python
area_fig = px.area(data,
                   x = 'Month',
                   y = ['Python Worldwide(%)', 'JavaScript Worldwide(%)',
                                     'Java Worldwide(%)', 'C# Worldwide(%)', 'PhP Worldwide(%)'],
                   title = 'Cumulative Worldwide Popularity of Programming Languages')
area_fig.show()
```

```python
violin_fig = px.violin(data.melt(id_vars = 'Month', var_name = 'Language', value_name = 'Popularity'),
                       x = 'Language', y = 'Popularity', box = True, points = 'all',
                       title = 'Density and Distribution of Programming Languages Popularity')
violin_fig.show()
```

```python
bubble_fig = px.scatter(data,
                        x = 'Python Worldwide(%)',
                        y = 'JavaScript Worldwide(%)',
                        size = 'Java Worldwide(%)',
                        title = 'Python vs JavaScript vs Java Popularity')
bubble_fig.show()
```

```python
pie_fig = px.pie(data.mean(numeric_only = True).reset_index(),
                 names = 'index',
                 values = 0,
                 title = 'Share of Worldwide Popularity Among Programming Languages')
pie_fig.show()
```

```python
s = data.select_dtypes(include = 'number').corr()
fig = px.density_heatmap(s, title = 'Correlation Heatmap')
fig.update_layout(
height = 600,
width = 900
)
fig.show()
```
